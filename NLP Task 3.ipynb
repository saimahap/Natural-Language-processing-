```python
def min_edit_distance(s1, s2):
    m, n = len(s1), len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]

    for i in range(m + 1):
        dp[i][0] = i
    for j in range(n + 1):
        dp[0][j] = j

    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]
            else:
                dp[i][j] = 1 + min(dp[i][j - 1], dp[i - 1][j], dp[i - 1][j - 1])

    return dp[m][n], dp
```

```python
def print_table(s1, s2, dp):
    print("      " + "  ".join(list(" " + s2)))
    for i, row in enumerate(dp):
        label = " " if i == 0 else s1[i - 1]
        print(f"{label}  " + "  ".join(str(x) for x in row))
```

```python
def backtrack_operations(s1, s2, dp):
    i, j = len(s1), len(s2)
    operations = []

    while i > 0 or j > 0:
        if i > 0 and j > 0 and s1[i - 1] == s2[j - 1]:
            i -= 1
            j -= 1
        elif i > 0 and j > 0 and dp[i][j] == dp[i - 1][j - 1] + 1:
            operations.append(f"Replace '{s1[i-1]}' at position {i-1} with '{s2[j-1]}'")
            i -= 1
            j -= 1
        elif i > 0 and dp[i][j] == dp[i - 1][j] + 1:
            operations.append(f"Remove '{s1[i-1]}' at position {i-1}")
            i -= 1
        elif j > 0 and dp[i][j] == dp[i][j - 1] + 1:
            operations.append(f"Insert '{s2[j-1]}' at position {i}")
            j -= 1

    operations.reverse()
    return operations
```

```python
s1 = "abcd"
s2 = "bcfe"

distance, table = min_edit_distance(s1, s2)
print(f"Minimum Edit Distance between '{s1}' and '{s2}': {distance}\n")
print_table(s1, s2, table)

print("\nSequence of operations:")
for op in backtrack_operations(s1, s2, table):
    print("-", op)
```
